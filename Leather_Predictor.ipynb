```python
image_width = 480
image_height = 480
```

```python
from keras.models import load_model

# Specify the path to the h5 model file
model_path = 'C:/Users/USER/Documents/Python Projects/Leather defect detection/Leather Detection/leather_detector.h5'
model = load_model(model_path)
```

```python
def preprocess_image(image):
    image = cv2.resize(image, (image_width, image_height))
    image = image.astype('float32') / 255.0
    return image

def predict_class(image, model):
    preprocessed_image = preprocess_image(image)
    input_image = np.expand_dims(preprocessed_image, axis=0)
    probabilities = model.predict(input_image)[0]
    predicted_class_index = np.argmax(probabilities)
    max_probability = np.max(probabilities)
    return predicted_class_index, max_probability
```

```python
# Define the class labels
class_labels = ['Leather', 'Not a Leather']

# Initialize the video capture
cap = cv2.VideoCapture(0)

while True:
    # Read the frame from the camera
    ret, frame = cap.read()
    
    if ret:
        # Preprocess the frame
        frame = cv2.resize(frame, (480, 480))
        frame = frame.astype(np.uint8)
        frame = frame / 255.0
        frame = np.expand_dims(frame, axis=0)
        
        # Make predictions
        predictions = model.predict(frame)
        class_index = np.argmax(predictions)
        class_label = class_labels[class_index]
        percentage = predictions[0][class_index] * 100
        
        # Draw rectangle around the leather piece
        # Replace these values with your specific coordinates
        x1, y1, x2, y2 = 100, 100, 300, 300
        frame = frame * 255
        frame = frame.astype(np.uint8)
        
        # Display predicted class and percentage
        label = f'{class_label}: {percentage:.2f}%'
        cv2.putText(frame[0], label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        print(label)
        
        # Display the frame
        cv2.imshow('Leather Piece', frame[0])
        
    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all windows
cap.release()
cv2.destroyAllWindows()
```
